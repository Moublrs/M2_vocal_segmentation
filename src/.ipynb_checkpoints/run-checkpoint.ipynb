# INFERENCE ON NPY SPECTROGRAMS


```python
import torch
import numpy as np
import librosa
import soundfile as sf
from pathlib import Path

from model import UNet


# =========================
# Paths
# =========================
TRACK_DIR = Path.home() / "TP SON UNET" / "data" / "spec_data_linear" / "test" / "Little Chicago's Finest - My Own_mixture"
MODEL_PATH = "unet_final.pth"
OUT_WAV = "vocals_pred.wav"


# =========================
# Params (match training)
# =========================
SR = 8192
N_FFT = 1024
HOP_LENGTH = 768

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

FRAME_SIZE = 128
STRIDE = 1   # 1 = tr√®s lent ; 32/64 = test rapide


# =========================
# Load model
# =========================
model = UNet().to(DEVICE)
model.load_state_dict(torch.load(MODEL_PATH, map_location=DEVICE))
model.eval()


# =========================
# Load spectrogram data (513 bins on disk)
# =========================
mix_mag_513 = np.load(TRACK_DIR / "mixture_linear.npy").astype(np.float32)      # (513, T) normalized
mix_phase_513 = np.load(TRACK_DIR / "mixture_phase.npy").astype(np.float32)    # (513, T)
mix_max = np.load(TRACK_DIR / "mix_max.npy").astype(np.float32)                # scalar

F513, T = mix_mag_513.shape
assert F513 == 513, f"Expected 513 freq bins on disk, got {F513}"

# =========================
# Drop Nyquist for the model -> 512 bins (as in the paper diagram)
# =========================
mix_mag_512 = mix_mag_513[:-1, :]      # (512, T)
mix_phase_512 = mix_phase_513[:-1, :]  # (512, T) (used only inside patches if needed)

F, T = mix_mag_512.shape
assert F == 512, f"Expected 512 bins for the model, got {F}"


# =========================
# Patch-based inference (overlap-add) in 512-bin space
# =========================
voc_norm_512 = np.zeros((F, T), dtype=np.float32)
weight = np.zeros((F, T), dtype=np.float32)

with torch.no_grad():
    for t0 in range(0, T - FRAME_SIZE + 1, STRIDE):
        mix_patch = mix_mag_512[:, t0:t0 + FRAME_SIZE]  # (512, 128)
        mix_t = torch.from_numpy(mix_patch).unsqueeze(0).unsqueeze(0).to(DEVICE)  # (1,1,512,128)

        mask_patch = model(mix_t).squeeze().cpu().numpy().astype(np.float32)      # (512,128)
        voc_patch = mask_patch * mix_patch                                        # (512,128)

        voc_norm_512[:, t0:t0 + FRAME_SIZE] += voc_patch
        weight[:, t0:t0 + FRAME_SIZE] += 1.0

# Handle tail if STRIDE doesn't land exactly on the last window
last_start = T - FRAME_SIZE
if last_start > 0 and (T - FRAME_SIZE) % STRIDE != 0:
    with torch.no_grad():
        mix_patch = mix_mag_512[:, last_start:last_start + FRAME_SIZE]
        mix_t = torch.from_numpy(mix_patch).unsqueeze(0).unsqueeze(0).to(DEVICE)

        mask_patch = model(mix_t).squeeze().cpu().numpy().astype(np.float32)
        voc_patch = mask_patch * mix_patch

        voc_norm_512[:, last_start:last_start + FRAME_SIZE] += voc_patch
        weight[:, last_start:last_start + FRAME_SIZE] += 1.0

voc_norm_512 /= np.maximum(weight, 1.0)

# De-normalize back to original magnitude scale
voc_mag_512 = voc_norm_512 * mix_max


# =========================
# Restore Nyquist bin (513th row) with zeros, then ISTFT with n_fft=1024
# =========================
nyquist_row = np.zeros((1, T), dtype=np.float32)
voc_mag_513 = np.vstack([voc_mag_512, nyquist_row])  # (513, T)

# Complex STFT using MIX phase (full 513 bins)
voc_stft_513 = voc_mag_513 * np.exp(1j * mix_phase_513)

# ISTFT with the original FFT size
voc_audio = librosa.istft(
    voc_stft_513,
    n_fft=N_FFT,
    hop_length=HOP_LENGTH,
    win_length=N_FFT
)

sf.write(OUT_WAV, voc_audio, SR)
print("Saved:", OUT_WAV)
```

# INFERENCE ON MP3/WAV SONG


```python
import torch
import numpy as np
import librosa
import soundfile as sf
from pathlib import Path

from model import UNet


# =========================
# Paths (edit these)
# =========================
MP3_PATH = Path("angels.wav")     # mixture audio (mp3)
MODEL_PATH = Path("unet_final.pth")
OUT_WAV = Path("vocals_pred.wav")


# =========================
# Params (must match training)
# =========================
SR = 8192
N_FFT = 1024
HOP_LENGTH = 768

FRAME_SIZE = 128          # model patch width in frames
STRIDE_FRAMES = 1        # overlap-add stride in frames (32/64 good for speed)

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"


# =========================
# Load model
# =========================
model = UNet().to(DEVICE)
model.load_state_dict(torch.load(MODEL_PATH, map_location=DEVICE))
model.eval()


# =========================
# Load mp3 -> STFT (513 bins)
# =========================
y, sr = librosa.load(str(MP3_PATH), mono=True)

stft_mix = librosa.stft(y, n_fft=N_FFT, hop_length=HOP_LENGTH)
mix_mag_513 = np.abs(stft_mix).astype(np.float32)          # (513, T)
mix_phase_513 = np.angle(stft_mix).astype(np.float32)      # (513, T)

mix_max = float(mix_mag_513.max() + 1e-8)                  # avoid div by 0
mix_mag_513_norm = (mix_mag_513 / mix_max).astype(np.float32)


# =========================
# Drop Nyquist bin for the model (512 bins)
# =========================
mix_mag_512 = mix_mag_513_norm[:-1, :]                     # (512, T)
F, T = mix_mag_512.shape
assert F == 512, f"Expected 512 bins for model input, got {F}"


# =========================
# Patch-based inference (overlap-add) in 512-bin space
# =========================
voc_norm_512 = np.zeros((F, T), dtype=np.float32)
weight = np.zeros((F, T), dtype=np.float32)

with torch.no_grad():
    for t0 in range(0, T - FRAME_SIZE + 1, STRIDE_FRAMES):
        mix_patch = mix_mag_512[:, t0:t0 + FRAME_SIZE]  # (512, 128)
        mix_t = torch.from_numpy(mix_patch).unsqueeze(0).unsqueeze(0).to(DEVICE)  # (1,1,512,128)

        mask_patch = model(mix_t).squeeze().cpu().numpy().astype(np.float32)      # (512,128)
        voc_patch = mask_patch * mix_patch                                        # (512,128)

        voc_norm_512[:, t0:t0 + FRAME_SIZE] += voc_patch
        weight[:, t0:t0 + FRAME_SIZE] += 1.0



voc_norm_512 /= np.maximum(weight, 1.0)


# =========================
# De-normalize + restore Nyquist bin + ISTFT (n_fft=1024)
# =========================
voc_mag_512 = voc_norm_512 * mix_max                        # back to original scale

nyquist_row = np.zeros((1, T), dtype=np.float32)
voc_mag_513 = np.vstack([voc_mag_512, nyquist_row])         # (513, T)

voc_stft_513 = voc_mag_513 * np.exp(1j * mix_phase_513)

voc_audio = librosa.istft(
    voc_stft_513,
    n_fft=N_FFT,
    hop_length=HOP_LENGTH,
    win_length=N_FFT,
    length=len(y)   # keep same length as input
)

sf.write(str(OUT_WAV), voc_audio, sr)
print("Saved:", OUT_WAV)
```

```python
import yt_dlp
import os
import logging
from pathlib import Path
from typing import Optional, Dict, Any

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
logger = logging.getLogger(__name__)


class YouTubeMP3Downloader:
    """
    A robust YouTube audio downloader with support for videos and playlists.
    """
    
    def __init__(self, output_path: str = "downloads", audio_quality: str = "192"):
        """
        Initialize the downloader.
        
        Parameters:
        - output_path: Directory to save files
        - audio_quality: Audio bitrate (128, 192, 256, 320)
        """
        self.output_path = Path(output_path)
        self.output_path.mkdir(parents=True, exist_ok=True)
        self.audio_quality = audio_quality
        self._validate_audio_quality()
    
    def _validate_audio_quality(self):
        """Validate that audio quality is within acceptable range."""
        valid_qualities = {"128", "192", "256", "320"}
        if self.audio_quality not in valid_qualities:
            logger.warning(f"Invalid quality '{self.audio_quality}'. Using default 192 kbps.")
            self.audio_quality = "192"
    
    def _check_ffmpeg(self) -> bool:
        """Check if FFmpeg is installed and accessible."""
        result = os.system("ffmpeg -version > /dev/null 2>&1")
        return result == 0
    
    def _get_ydl_opts(self, include_uploader: bool = False) -> Dict[str, Any]:
        """
        Generate yt-dlp options.
        
        Parameters:
        - include_uploader: Whether to include uploader name in filename
        """
        # Base template
        if include_uploader:
            outtmpl = os.path.join(str(self.output_path), '%(title)s - %(uploader)s.%(ext)s')
        else:
            outtmpl = os.path.join(str(self.output_path), '%(title)s.%(ext)s')
        
        return {
            'format': 'bestaudio/best',
            'postprocessors': [{
                'key': 'FFmpegExtractAudio',
                'preferredcodec': 'mp3',
                'preferredquality': self.audio_quality,
            }],
            'outtmpl': outtmpl,
            'quiet': False,
            'no_warnings': False,
            'progress_hooks': [self._progress_hook],
            'socket_timeout': 30,
            'retries': 3,
            'fragment_retries': 3,
        }
    
    def _progress_hook(self, d):
        """Handle download progress updates."""
        if d['status'] == 'downloading':
            percent = d.get('_percent_str', 'N/A')
            speed = d.get('_speed_str', 'N/A')
            eta = d.get('_eta_str', 'N/A')
            logger.info(f"Downloading: {percent} at {speed} (ETA: {eta})")
        elif d['status'] == 'finished':
            logger.info("Download finished. Processing audio...")
        elif d['status'] == 'error':
            logger.error(f"Download error: {d.get('error', 'Unknown error')}")
    
    def download(self, url: str, include_uploader: bool = False) -> bool:
        """
        Download audio from YouTube URL.
        
        Parameters:
        - url: YouTube video or playlist URL
        - include_uploader: Include uploader name in filename
        
        Returns:
        - True if successful, False otherwise
        """
        # Validate FFmpeg
        if not self._check_ffmpeg():
            logger.error("FFmpeg is not installed or not in PATH. Please install FFmpeg.")
            logger.info("Visit: https://ffmpeg.org/download.html")
            return False
        
        # Validate URL
        if not self._validate_url(url):
            logger.error(f"Invalid YouTube URL: {url}")
            return False
        
        ydl_opts = self._get_ydl_opts(include_uploader=include_uploader)
        
        try:
            logger.info(f"Starting download from: {url}")
            with yt_dlp.YoutubeDL(ydl_opts) as ydl:
                info = ydl.extract_info(url, download=True)
                
                # Check if it's a playlist
                if 'entries' in info:
                    logger.info(f"Successfully downloaded {len(info['entries'])} items from playlist")
                else:
                    logger.info(f"Successfully downloaded: {info.get('title', 'Unknown')}")
            
            logger.info(f"Files saved to: {self.output_path.absolute()}")
            return True
        
        except yt_dlp.utils.DownloadError as e:
            logger.error(f"Download error: {e}")
            return False
        except Exception as e:
            logger.error(f"Unexpected error: {type(e).__name__}: {e}")
            return False
    
    @staticmethod
    def _validate_url(url: str) -> bool:
        """Validate if URL is a valid YouTube URL."""
        youtube_domains = ('youtube.com', 'youtu.be', 'youtube-nocookie.com')
        return any(domain in url for domain in youtube_domains)
    
    def get_video_info(self, url: str) -> Optional[Dict[str, Any]]:
        """
        Extract information about a YouTube video without downloading.
        
        Returns:
        - Dictionary with video info or None if extraction fails
        """
        try:
            with yt_dlp.YoutubeDL({'quiet': True}) as ydl:
                info = ydl.extract_info(url, download=False)
                return {
                    'title': info.get('title'),
                    'uploader': info.get('uploader'),
                    'duration': info.get('duration'),
                    'view_count': info.get('view_count'),
                    'is_playlist': 'entries' in info,
                    'items_count': len(info.get('entries', [])) if 'entries' in info else 1,
                }
        except Exception as e:
            logger.error(f"Failed to extract video info: {e}")
            return None


# ========== USAGE EXAMPLES ==========
if __name__ == "__main__":
    # Initialize downloader with 192 kbps quality
    downloader = YouTubeMP3Downloader(output_path="songs_ytb", audio_quality="192")
    

    
    # Example 2: Playlist
    playlist_url = "https://www.youtube.com/watch?v=8UVNT4wvIGY&list=PLgDP5UKLWaCA_cJeIJBcVLbLFid49AMQh"
    downloader.download(playlist_url, include_uploader=True)
    
```

# TRAINING ON SONGS FROM YOUTUBE

```python
import torch
import torchaudio
from demucs.pretrained import get_model
from demucs.apply import apply_model
from pathlib import Path

TRACK_DIR = Path.home() / "TP SON UNET" / "src" / "songs_ytb"
VOCAL_DIR = Path.home() / "TP SON UNET" / "src" / "vocals"

VOCAL_DIR.mkdir(parents=True, exist_ok=True)

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Device: {device}")

# Load model once
print("Loading model...")
model = get_model("mdx_extra").to(device)  # Faster than htdemucs_ft
model.eval()

# Get all MP3 files
mp3_files = list(TRACK_DIR.glob("*.mp3"))
print(f"Found {len(mp3_files)} MP3 files\n")

for i, mp3_path in enumerate(mp3_files, 1):
    print(f"[{i}/{len(mp3_files)}] Processing: {mp3_path.name}")
    
    try:
        # Load audio
        waveform, sr = torchaudio.load(str(mp3_path))
        waveform = waveform.to(device)
        waveform = waveform.unsqueeze(0)  # Add batch dimension
        
        # Separate
        with torch.no_grad():
            stems = apply_model(model, waveform, device=device, shifts=0, overlap=0.1)
        
        stems = stems.squeeze(0)
        
        # Save vocals (index 3)
        vocals = stems[3]
        output_path = VOCAL_DIR / f"{mp3_path.stem}_vocal.mp3"
        torchaudio.save(str(output_path), vocals.cpu(), sr)
        print(f"  ‚úì Saved: {output_path.name}\n")
    
    except Exception as e:
        print(f"  ‚úó Error: {e}\n")

print(f"Done! Vocals saved to: {VOCAL_DIR}")
```

```python
import numpy as np
import librosa
import torch
from pathlib import Path
from mir_eval.separation import bss_eval_sources
from model import UNet

SR = 8192
N_FFT = 1024
HOP_LENGTH = 768
FRAME_SIZE = 128
STRIDE = 64
DEVICE = torch.device("cuda:0")

model = UNet().to(DEVICE)
model.load_state_dict(torch.load("./models/unet_final_fine_tuned.pth", map_location=DEVICE, weights_only=False))
model.eval()

def separate(mix_path):
    y, _ = librosa.load(mix_path, sr=SR, mono=True)
    stft = librosa.stft(y, n_fft=N_FFT, hop_length=HOP_LENGTH)
    mag, phase = np.abs(stft).astype(np.float32), np.angle(stft)
    
    mx = mag.max() + 1e-8
    mag_norm = (mag / mx)[:-1, :]
    F, T = mag_norm.shape
    
    voc = np.zeros((F, T), dtype=np.float32)
    w = np.zeros((F, T), dtype=np.float32)
    
    with torch.no_grad():
        for t in range(0, T - FRAME_SIZE + 1, STRIDE):
            p = mag_norm[:, t:t+FRAME_SIZE]
            m = model(torch.from_numpy(p).unsqueeze(0).unsqueeze(0).to(DEVICE)).squeeze().cpu().numpy()
            voc[:, t:t+FRAME_SIZE] += m * p
            w[:, t:t+FRAME_SIZE] += 1.0
    
    voc /= np.maximum(w, 1.0)
    voc_mag = np.vstack([voc * mx, np.zeros((1, T))])
    return librosa.istft(voc_mag * np.exp(1j * phase), n_fft=N_FFT, hop_length=HOP_LENGTH, length=len(y))

def evaluate(est, ref):
    L = min(len(est), len(ref))
    sdr, sir, sar, _ = bss_eval_sources(ref[:L].reshape(1, -1), est[:L].reshape(1, -1))
    return sdr[0], sir[0], sar[0]

data_dir = Path("../downloads")
results = []

for voc_path in sorted(data_dir.glob("*_vocal.mp3")):
    mix_path = data_dir / voc_path.name.replace("_vocal.mp3", ".mp3")
    if not mix_path.exists():
        continue
    
    name = voc_path.stem.replace("_vocal", "")
    
    ref, _ = librosa.load(voc_path, sr=SR, mono=True)
    est = separate(mix_path)
    
    sdr, sir, sar = evaluate(est, ref)
    results.append((name, sdr, sir, sar))
    print(f"{name}: SDR={sdr:.2f} SIR={sir:.2f} SAR={sar:.2f}")

print(f"\nMoyenne: SDR={np.mean([r[1] for r in results]):.2f} SIR={np.mean([r[2] for r in results]):.2f} SAR={np.mean([r[3] for r in results]):.2f}")
```
